# **Gradient Descent**

**Function**

```python
def linear(X, Y, alpha, iteration):

  m = Y.size # training set length

  theta = np.zeros( (X.shape[1] , 1) ) # initializing theta

  cost_list = [] # to store cost Function

  # mean normalization

  for i in range( 1, X.shape[1] ):

     mean = np.mean( X[:,i])

     deviate = ( np.amax( X[:,i] ) - np.amin( X[:,i]))

     for j in range(X.shape[0]):

       X[j,i] = ( X[j,i] - mean ) / deviate


  for k in range (Y.size):

    Y[k,0] =  ( Y[k,0] - np.mean( Y ) ) / ( np.amax( Y ) - np.amin( Y ))
  

  for z in range(iteration): 

   

      y_pred = np.dot(X, theta) # hypothesis

      cost = (1/(2*m))* np.sum(np.square(y_pred - Y)) # Cost Function

      d_cost = (1/m)* np.sum(np.dot(X.T, (y_pred - Y) )) # differentiating Cost Function w.r.t Theta

      temp = theta # storing theta in a temp variable

      temp = theta - alpha * d_cost # calculating new theta

      theta = temp # updating theta

      cost_list.append(cost) # updating cost Function list


      # to see if cost is decreasing

      if z % 1000 == 0:
        print("Cost is", cost)

      

  return theta , cost_list
```

**Running Function**

```python
iteration = 10000

alpha = 0.001

theta , cost_list = linear(X, Y, alpha = alpha, iteration = iteration)
```

**Plotting Cost vs Iteration Graph**

```python
iter = np.arange( 0, iteration)
plt.plot(iter, cost_list)
plt.show()
```
