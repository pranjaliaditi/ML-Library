# Choosing Random Centroid

```python
# Co-ordinates of Centroid = (X1, X2 ... X784)

# we have to choose 19 centroids (because 19 clusters)

# for that we will randomly select 19 rows of X_test
```

```python
np.random.shuffle(X_test)

Centroid = X_test[0:19,:]

pd.DataFrame(Centroid)
```

```python
# calculating distance of a point from all the Centroid points

def distance(i):

  dist_list = []

  j = 0

  for j in range(Centroid.shape[0]):

    dist = np.sqrt(np.sum((X_test[i,:] - Centroid[j,:]) ** 2 ))

    dist_list.append(dist)


  return dist_list
```

```python
i = 0

cluster = []

# calculating for all 14800 points

for i in range(X_test.shape[0]):

  distance_list = distance(i=i)

  min_value = np.min(distance_list)

  min_dist_index = distance_list.index( min_value  )

  Centroid_new = []

  cluster.append( (min_dist_index + 1) )

  Centroid_new.append(distance_list) #updating new Centroid 19 X 784 (we will do till no change in centroid)

print(cluster)
```

```python
#Cluster Sorting Formed

Number = np.zeros((K,1))

i = 0

for i in range(K):

  Number[i,] = cluster.count(i+1)

pd.DataFrame(Number)
```
